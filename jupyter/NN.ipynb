```python
import numpy as np 
import pandas as pd 
from pathlib import Path
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
PATH = Path(r'C:\Users\niti.mishra\Documents\2_TDMDAL\projects\covid_tests\covid_tests')
# PATH = Path(r'/Users/peaceforlives/Documents/Projects/covid_symptoms')
DATA = Path.joinpath(PATH, 'data')
RESULTS = Path.joinpath(PATH, 'results')
```

```python
covid_data = pd.read_csv(Path.joinpath(DATA, 'RawData_IMOH_Sep_8.csv')) 
covid_data.set_index('Validation', inplace=True)
covid_data.head()
```

```python
cols = ['cough', 'fever', 'sore_throat', 'shortness_of_breath', 'head_ache', 'contact', 'abroad']
X = covid_data[cols].copy()
y = covid_data['testresult'].copy().to_numpy().reshape(-1,1)
```

```python
train = covid_data.index.get_loc('Training')
valid = covid_data.index.get_loc('Validation')
test = covid_data.index.get_loc('Test')

X_train = X[train]
y_train = y[train]
print(len(y_train))

X_valid = X[valid]
y_valid = y[valid]
print(len(y_valid))

X_test = X[test]
y_test = y[test]
print(len(y_test))
```

```python
# create a sequential model and add one layer at a time
# fully connected layer (dense)
model = Sequential()
model.add(Dense(6, input_dim=X.shape[1], activation='relu')) # first layer
model.add(Dense(3, activation='relu'))                       # second layer
model.add(Dense(1, activation='sigmoid'))                    # output layer
```

```python
# Compiling the model uses the efficient numerical libraries under the covers 
# the so-called backend) such as Theano or TensorFlow. 

# training a network = finding the best set of weights to map inputs to outputs
# specify the loss function to use to evaluate a set of weights (cross entropy)
# specify optimizer to search through different weights stochastic gradient descent algorithm

# compile 
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# Epoch: One pass through all of the rows in the training dataset, will run for fixed no of iterations
# Batch: One or more samples considered by the model within an epoch before weights are updated

# fit the keras model on the dataset
model.fit(X_train, y_train, epochs=150, batch_size=10,  validation_data=(X_valid, y_valid))
```

```python
# evaluate the keras model
loss, accuracy = model.evaluate(X, y)
print("\nLoss: %.2f, Accuracy: %.2f%%" % (loss, accuracy*100))
```

```python
y_pred = model.predict_classes(X_test)
y_pred_prob = model.predict_proba(X_test)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, auc, roc_auc_score
```

```python
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, average='weighted')
recall = recall_score(y_test, y_pred, average='weighted')
# auc = roc_auc_score(y_test, y_pred, average='weighted')
auc = roc_auc_score(y_test, y_pred_prob, average='weighted')
f1_test = f1_score(y_test, y_pred, average='weighted')
```

```python
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import GridSearchCV
```

```python
# Function to create model, required for KerasClassifier
def create_model():
    # create model
    model = Sequential()
    model.add(Dense(6, input_dim=X.shape[1], activation='relu')) # first layer
    model.add(Dense(3, activation='relu'))                       # second layer
    model.add(Dense(1, activation='sigmoid'))                    # output layer
    # Compile model
    model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
    return model
```

```python
# fix random seed for reproducibility
np.random.seed(42)
# create model
model = KerasClassifier(build_fn=create_model, verbose=0)

# define the grid search parameters
batch_size = [10, 20, 40, 60, 80, 100]
epochs = [10, 50, 100]

param_grid = dict(batch_size=batch_size, epochs=epochs)
grid = GridSearchCV(estimator=model, param_grid=param_grid,cv=3)
```

```python
grid_result = grid.fit(X_train, y_train)

# summarize results
print("Best: %f using %s" % (grid_result.best_score_, grid_result.best_params_))
means = grid_result.cv_results_['mean_test_score']
stds = grid_result.cv_results_['std_test_score']
params = grid_result.cv_results_['params']
for mean, stdev, param in zip(means, stds, params):
    print("%f (%f) with: %r" % (mean, stdev, param))
```
